```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("iris.csv")
print(df.shape)      
print(df.columns)  
df.head()
col=df["sepal_length"]
mean_val=col.mean()
med_val=col.median()
mode_val=col.mode()
print(mean_val,mode_val,med_val)
var=col.var()
std=col.std()
q1=col.quantile(0.25)
q2=col.quantile(0.50)
q3=col.quantile(0.75)

iqr=q3-q1


import matplotlib.pyplot as plt

col = df["petal_length"]     # single column, single brackets
mean_val = col.mean()
median_val = col.median()
mode_val = col.mode()[0]

print(mean_val, median_val, mode_val)   # should print 3 plain numbers
import matplotlib.pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(12, 4.5))

# Histogram with mean/median/mode
axes[0].hist(col, bins=15, color="skyblue", edgecolor="black", zorder=1)
axes[0].axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean={mean_val:.2f}", zorder=5)
axes[0].axvline(median_val, color="green", linestyle="-.", linewidth=2, label=f"Median={median_val:.2f}", zorder=5)
axes[0].axvline(mode_val, color="orange", linestyle=":", linewidth=2, label=f"Mode={mode_val:.2f}", zorder=5)
axes[0].axvline(var, color="violet", linestyle="solid", linewidth=2, label=f"Variance={var:.2f}", zorder=5)
axes[0].axvline(std, color="pink", linestyle="solid", linewidth=2, label=f"standard deviation={std:.2f}", zorder=5)
axes[0].set_xlabel("petal_length")

axes[0].set_ylabel("Frequency")
axes[0].set_title("Histogram of petal_length")
axes[0].legend(fontsize=8)

# Boxplot with quartiles/IQR
axes[1].boxplot(col, vert=True, patch_artist=True, boxprops=dict(facecolor="lightblue"))
axes[1].set_xticks([1])
axes[1].set_xticklabels(["petal_length"])
axes[1].set_ylabel("petal_length")
axes[1].set_title(f"Boxplot of petal_length\nQ1={q1:.2f}, Q3={q3:.2f}, IQR={iqr:.2f}")

plt.tight_layout()
plt.show()
```
